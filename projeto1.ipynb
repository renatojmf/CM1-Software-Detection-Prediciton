# Crawler


```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
from time import sleep
sites = {
    'webmotors':'https://www.webmotors.com.br/',
    'autoline':'https://www.autoline.com.br/',
    'meucarronovo':'https://www.meucarronovo.com.br/',
    'icarros':'https://www.icarros.com.br/',
    'mobialto':'https://www.mobiauto.com.br/',
    'vrum':'https://diariodepernambuco.vrum.com.br/',
    'usados':'https://www.usadosbr.com/',
    'seminovos':'https://seminovos.com.br/',
    'olx':'https://pe.olx.com.br/autos-e-pecas/carros-vans-e-utilitarios?q=carro',
    'mercadolivre':'https://www.mercadolivre.com.br/'
}

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

def get_website_links(url):
  urls = []
  r = requests.get(url,headers=header)
  soup = BeautifulSoup(r.content,'html.parser')
  for _a in soup.find_all('a'):
    _href = _a.attrs.get('href')
    if _href == "" or _href is None:
      continue
    urls.append(_href)
  return urls,soup

def is_valid(url):
  parsed = urlparse(url)
  return bool(parsed.netloc) and bool(parsed.scheme)

def check_urls(urls):
  new_url = []
  for url in urls:
    if is_valid(url):
      new_url.append(url)
  return new_url

def myCrawl(url,number_visits):
  _links = []
  _links_visited = []
  _links.append(url)

  while number_visits > 0:
    link_to_visit = _links.pop(0)
    if  link_to_visit in _links_visited:
      continue
    new_links, new_page = get_website_links(link_to_visit)
    new_links = check_urls(new_links)
    _links_visited.append((link_to_visit,new_page))
    _links.extend(new_links)
    number_visits -=1
  return _links_visited

if __name__ == '__main__':
  list = myCrawl(sites['olx'],50)
  for link,page in list:
    print(link)
```

```python
import nltk  
nltk.download('punkt')
import numpy as np  
import random  
import string

import bs4 as bs  
import urllib.request  
import re  
```

```python
from urllib.request import Request, urlopen
url = 'https://www.usadosbr.com/carros-e-utilitarios/honda'
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'}
request = Request(url, headers=headers)
raw_html = urlopen(request).read()

#raw_html = urllib.request.urlopen('https://www.usadosbr.com/carros-e-utilitarios/honda',headers=hdr).content  
#raw_html = raw_html.read()

article_html = bs.BeautifulSoup(raw_html, 'lxml')

article_paragraphs = article_html.find_all('p')
article_as = article_html.find_all('a')

article_text = ''

for para in article_paragraphs:  
    article_text += para.text
for a in article_as:
  article_text += a.text
corpus = nltk.sent_tokenize(article_text)
```

```python
for i in range(len(corpus )):
    corpus [i] = corpus [i].lower()
    corpus [i] = re.sub(r'\W',' ',corpus [i])
    corpus [i] = re.sub(r'\s+',' ',corpus [i])

print(len(corpus))
for i in range(len(corpus )):
  print(corpus[i])
#print(corpus[10])
```

```python
wordfreq = {}
for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
```

```python
import heapq
most_freq = heapq.nlargest(200, wordfreq, key=wordfreq.get)
```

```python
sentence_vectors = []
for sentence in corpus:
    sentence_tokens = nltk.word_tokenize(sentence)
    sent_vec = []
    for token in most_freq:
        if token in sentence_tokens:
            sent_vec.append(1)
        else:
            sent_vec.append(0)
    sentence_vectors.append(sent_vec)
```

```python
sentence_vectors = np.asarray(sentence_vectors)
sentence_vectors
```

```python
word_idf_values = {}
for token in most_freq:
    doc_containing_word = 0
    for document in corpus:
        if token in nltk.word_tokenize(document):
            doc_containing_word += 1
    word_idf_values[token] = np.log(len(corpus)/(1 + doc_containing_word))
```

```python
word_tf_values = {}
for token in most_freq:
    sent_tf_vector = []
    for document in corpus:
        doc_freq = 0
        for word in nltk.word_tokenize(document):
            if token == word:
                  doc_freq += 1
        word_tf = doc_freq/len(nltk.word_tokenize(document))
        sent_tf_vector.append(word_tf)
    word_tf_values[token] = sent_tf_vector
```

```python
tfidf_values = []
for token in word_tf_values.keys():
    tfidf_sentences = []
    for tf_sentence in word_tf_values[token]:
        tf_idf_score = tf_sentence * word_idf_values[token]
        tfidf_sentences.append(tf_idf_score)
    tfidf_values.append(tfidf_sentences)
```

```python
tf_idf_model = np.asarray(tfidf_values)
tf_idf_model = np.transpose(tf_idf_model)
tf_idf_model
```
